```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
final_data = pd.read_csv("./final_data.csv")
final_data['date'] = pd.to_datetime(final_data['date'])
final_data.set_index(['age_group', 'date'], inplace=True)
final_data
```

```python
age_groups = final_data.index.get_level_values('age_group').unique().to_list()
age_groups
```

```python
# plot cum_case
plt.rcParams['figure.figsize'] = (9,6)
for age_group in age_groups:
    data = final_data.loc[age_group]['cum_case']
    plt.plot(data, label=age_group)

plt.grid()
plt.title("Cumulative cases in phase 2")
plt.legend()
plt.show()
```

```python
# plot cum_case
plt.rcParams['figure.figsize'] = (9,6)
for age_group in age_groups:
    data = final_data.loc[age_group]['cum_admission']
    plt.plot(data, label=age_group)

plt.grid()
plt.title("Cumulative hospitalisations in phase 2")
plt.legend()
plt.show()
```
